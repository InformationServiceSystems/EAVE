```python
df = pd.read_csv('energy_price_daily.csv')
```

```python
df.head()
```

```python
# Convert 'date' column to datetime
df['Datum von'] = pd.to_datetime(df['Datum von'])

# Extract the year from the 'date' column
df['year'] = df['Datum von'].dt.year

# Group by the year and calculate the average price for each year
avg_price_per_year = df.groupby('year')['DE/AT/LU [€/MWh] Berechnete Auflösungen'].mean().to_dict()

# Print the resulting dictionary
print(avg_price_per_year)
```
